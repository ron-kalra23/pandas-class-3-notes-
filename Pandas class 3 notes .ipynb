# date Time manipulation

```python
import pandas as pd 
import numpy as np
```

```python
date_range = pd.date_range('4/12/2021' , periods = 12 , freq= 'M')#start from date 04/12/2021 ,#frequnecy = range month wise 
dates = pd.DataFrame(date_range , columns = ['Date'])   #periods = no of dates we want #frequeny m shows monthswise.
#date_range = our data now we make it dataframe.# when freq is month it will take last date of particular month
```

```python
dates
```

```python
dates['Day'] = dates['Date'].dt.day # after = the dataframe dates have column date take days
dates['Months'] = dates['Date'].dt.month #take month values 
dates['Year'] = dates['Date'].dt.year# take year values #three new column for day , date , year 

dates
```

```python
dates.drop('Date' , axis = 1 , inplace = True ) #to drop the column from the dataframe 
```
